### WAP to calculate the price of pizza by using linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

diameter = np.array([6,8,10,12,14]).reshape(-1,1)
price =([13,16,19,22,25])

model = LinearRegression()
model.fit(diameter, price)
newDiameter = np.array([21]).reshape(-1,1)

newPrice = model.predict(newDiameter)
print("Price of pizza of diameter 21cm is ", newPrice)

plt.plot(diameter, price)
plt.title("Linear Regression")
plt.xlabel("Diameter")
plt.ylabel("Price")
plt.show()
```
